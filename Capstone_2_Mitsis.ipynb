```python
# Import the pandas library and read the Genotypes_capstone.csv file, using ';' as the delimiter.
import pandas as pd
genotypes = pd.read_csv("Genotypes_capstone.csv", sep=";")

# Define a function that takes a list of genotype values
# and returns a list of True/False indicating if each value is greater than 0.
def find_present_nucleotides(nucs_in_population):
    return [x > 0 for x in (nucs_in_population)]

def is_biallelic(row):
    # Call the 'find_present_nucleotides' function on specific columns for every population group
    # to create lists of True/False values.
    # Combine the True/False lists using the 'and' operator to create a total presence list.
    nuc_presence_total = [any(items) for items in 
                          zip(find_present_nucleotides(row[2:6]), 
                              find_present_nucleotides(row[6:10]), 
                              find_present_nucleotides(row[10:14]), 
                              find_present_nucleotides(row[14:18]))]
    return sum(nuc_presence_total) == 2


# Filter the genotypes to include only biallelic SNPs.
filtered_genotypes = genotypes[genotypes.apply(is_biallelic, axis=1)]
filtered_genotypes
```
